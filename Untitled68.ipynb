DATA PRE PROCESSING

Encoding the categorical features 

```python
# Encoding 'sex' column
df.replace({'sex': {'male': 0, 'female': 1}}, inplace=True)

# Encoding 'smoker' column
df.replace({'smoker': {'yes': 1, 'no': 0}}, inplace=True)  # Changed values to match the description (yes = 1, no = 0)

# Encoding 'region' column
df.replace({'region': {'southeast': 0, 'southwest': 1, 'northeast': 2, 'northwest': 3}}, inplace=True)
```

Splitting  the features and  Target

```python
# Splitting the DataFrame into features (X) and target (Y)
X = df.drop(columns='charges', axis=1)
Y = df['charges']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and testing data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2,random_state=2)
```

Model Training


Linear Regression

```python
#Loading the Linear Regression model
regressor = LinearRegression()
```

```python
regressor.fit(X_train, Y_train)
```

Model Evaluation 

```python
#prediction on traing data
training_data_prediction=regressor.predict(X_train)
```

```python
#R squared value
r2_train = metrics.r2_score(Y_train, training_data_prediction)
print('R squared value:', r2_train)
```

```python
#prediction on test data
test_data_prediction=regressor.predict(X_test)
```

```python
#R squared value
r2_test = metrics.r2_score(Y_test, test_data_prediction)
print('R squared value:', r2_test)
```

Building a Prediactive System

```python
# Example data with 'children' feature
data = {
    'age': [23, 45, 34],
    'sex': [1, 0, 1],
    'bmi': [22.0, 28.5, 25.6],
    'smoker': [0, 1, 0],
    'region': [0, 1, 2],
    'children': [2, 0, 1]  # Ensure 'children' is included
}
X = pd.DataFrame(data)
Y = pd.Series([3000, 4000, 3500])

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Define and train the model
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Input data for prediction with all required features
input_data = {
    'age': [46],
    'sex': [1],         # Assuming encoding where 1 represents male
    'bmi': [33.44],
    'smoker': [0],      # Assuming encoding where 0 represents non-smoker
    'region': [0],      # Assuming encoding with 0 for southeast
    'children': [1]     # Ensure 'children' feature is included
}

# Convert to DataFrame with correct column names and order
input_data_df = pd.DataFrame(input_data)

# Make prediction
prediction = regressor.predict(input_data_df)
print('Prediction:', prediction[0])

# Print the result with a formatted string
print(f'The insurance cost is USD {prediction[0]:,.2f}')
```
